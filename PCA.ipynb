# 数据降维

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv('data/trainData.csv')
data.head()
```

```python
features = ['HomePlanet', 'Destination', 'Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'VIP', 'CryoSleep', 'Deck', 'Room', 'Info']
X = data[features]
X
```

```python
X = X.fillna(X.median())

# Standardizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Applying PCA
pca = PCA(n_components=10)  # Let's reduce the data to 2 components for visualization purposes
X_pca = pca.fit_transform(X_scaled)

# Amount of variance explained by the two principal components
explained_variance = pca.explained_variance_ratio_

# Resulting transformed data and explained variance
X_pca, explained_variance
```
