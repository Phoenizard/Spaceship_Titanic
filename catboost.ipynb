## 使用Catboost进行分类

```python
import numpy as np
import pandas as pd

from catboost import CatBoostClassifier, Pool
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('data/trainData.csv')
test_data = pd.read_csv('data/testData.csv')

features = ['HomePlanet', 'Destination', 'Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'VIP', 'CryoSleep', 'Deck', 'Room', 'Info']
X, Y = data[features], data['Transported']

# Spilt the data into training and validation
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Create a pool for training and validation
train_pool = Pool(X_train, Y_train)
val_pool = Pool(X_val, Y_val)

# Create the model
model = CatBoostClassifier(iterations=1000, depth=6, learning_rate=0.1, loss_function='Logloss', verbose=True)

# Train the model
model.fit(train_pool, eval_set=val_pool)
```

```python
# Predict the test data
test_data['Transported'] = model.predict(test_data[features])

submission = test_data[['PassengerId', 'Transported']].copy()
submission.loc[:, 'Transported'] = submission['Transported'].apply(lambda x: True if x == 1 else False)
submission.to_csv('data/Submission.csv', index=False)
```
